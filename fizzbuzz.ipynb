__FIZZBUZZ PROJECT__

>this is a simple control flow project using range function, if statement and try-catch exemptions

- it accepts input from user of int number 

- if divisible by 3 and 5 it print fizzbuzz

- if only divivsible by just 3 it print fizz

- if only divisible by just 5 if print buzz

```python
def fizzbuzz(n):
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)

# Get user input and run FizzBuzz
try:
    num = int(input("Enter a number: "))
    fizzbuzz(num)
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
import time

# Function to implement the countdown timer with FizzBuzz logic
def countdown_timer(seconds):
    start_time = time.time()  # Record the start time
    
    try:
        while seconds > 0:
            elapsed = int(time.time() - start_time)  # Calculate elapsed time
            remaining = seconds - elapsed  # Calculate remaining time

            if remaining < 0:
                break  # Stop if time runs out

            mins_rem, secs_rem = divmod(remaining, 60)
            mins_elap, secs_elap = divmod(elapsed, 60)

            # FizzBuzz logic applied to remaining seconds
            if remaining % 3 == 0 and remaining % 5 == 0:
                fb_text = " FizzBuzz!"
            elif remaining % 3 == 0:
                fb_text = " Fizz!"
            elif remaining % 5 == 0:
                fb_text = " Buzz!"
            else:
                fb_text = ""

            # Display countdown time along with FizzBuzz message
            time_format = f'Elapsed: {mins_elap:02d}:{secs_elap:02d} | Remaining: {mins_rem:02d}:{secs_rem:02d}{fb_text}'
            print(time_format, end="\r")

            # Alert when 10 seconds or less remain
            if remaining <= 10:
                print(f"\nYou have {remaining} seconds remaining!", end="\r")

            time.sleep(1)  # Wait for 1 second

        print("\nTime's up!")  # Move to new line after countdown ends

    except ValueError:
        print("Invalid input. Please enter an integer.")

# Main function to get user input
def main():
    print("Countdown Timer with FizzBuzz")
    try:
        seconds_input = int(input("Enter time in seconds: "))
        countdown_timer(seconds_input)
    except ValueError:
        print("Invalid input. Please enter an integer.")

if __name__ == "__main__":
    main()
```
